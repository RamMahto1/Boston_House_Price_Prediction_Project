# Boston House Price Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("boston.csv")
df.head()
```

```python
# shape of dataset
df.shape
```

```python
# checking null values
df.isnull().sum().sum()
```

```python
# checking duplicate values
df.duplicated().sum()
```

```python
# number of unnique value in this dataset
df.nunique()
```

```python
# data information
df.info()
```

```python
# statatics test 
df.describe()
```

```python
numerical_feature=[feature for feature in df if df[feature].dtype!="O"]
categorical_feature = [feature for feature in df if df[feature].dtype=="O"]

print("we have {} numerical feature {}".format(len(numerical_feature),numerical_feature))
print("we have {} categorical feature {} ".format(len(categorical_feature),categorical_feature))
```

```python
numerical_features = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']]
numerical_features
```

```python
plt.figure(figsize=(18,6))
for i, feature in enumerate(numerical_features):
    plt.subplot(2,7, i+1)
    sns.histplot(df[feature], kde=True)
    plt.title(f"Histogram for: {feature}")
    plt.xlabel(feature)
plt.show()
```

```python
sns.heatmap(df)
```

```python
plt.figure(figsize=(18,6))
for i, feature in enumerate(numerical_features):
    plt.subplot(2,7, i+1)
    sns.distplot(df[feature], kde=True)
    plt.title(f"Distplot for: {feature}")
    plt.xlabel(feature)
plt.show()
```

```python
plt.figure(figsize=(18,6))
for i, feature in enumerate(numerical_features):
    plt.subplot(2,7, i+1)
    sns.violinplot(df[feature], kde=True)
    plt.title(f"violinplot for: {feature}")
    plt.xlabel(feature)
plt.show()

print("==================================================================================================================================")

plt.figure(figsize=(18,6))
for i, feature in enumerate(numerical_features):
    plt.subplot(2,7, i+1)
    sns.boxplot(df[feature])
    plt.title(f"violinplot for: {feature}")
    plt.xlabel(feature)
plt.show()
```
